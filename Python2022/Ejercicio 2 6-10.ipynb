```python
import pandas as pd
import matplotlib.pyplot as plt1

archivo = pd.read_csv("nombres-2010-2014.csv", sep=",")
tamaño = archivo["nombre"].size

def nombres2012():

    contador = 0

    for i in range(0, tamaño):
        anio = archivo.at[i, "anio"]

        if anio == 2012:
            contador = contador + 1
    
    print("En 2012 se usaron ", contador, " nombres")

def nombresConS2012():
    contador = 0

    for i in range(0, tamaño):
        anio = archivo.at[i, "anio"]
        nombre = archivo.at[i, "nombre"]

        if anio == 2012 and nombre[0] == "S":
            contador = contador + 1
    
    print("En 2012 se usaron ", contador, " nombres empezados en S")

def nombresConVocal():
    contador = 0

    for i in range(0, tamaño):
        nombre = archivo.at[i, "nombre"]

        if nombre[0] == "A" or nombre[0] == "E" or nombre[0] == "I" or nombre[0] == "O" or nombre[0] == "U":
            contador = contador + 1
    
    print("Se usaron ", contador, " nombres empezados en una vocal")

def nombresVocal20102013():
    contador = 0

    for i in range(0, tamaño):
        nombre = archivo.at[i, "nombre"]
        anio = archivo.at[i, "anio"]

        if anio > 2009 and anio < 2014:
            if nombre[0] == "A" or nombre[0] == "E" or nombre[0] == "I" or nombre[0] == "O" or nombre[0] == "U":
                contador = contador + 1

    print("Entre 2010 y 2013 se usaron ", contador, " nombres empezados en una vocal")

def nombreMasUsado():
    nombres = {}

def nombresConCadaVocal():
    contadorA = 0
    contadorE = 0
    contadorI = 0
    contadorO = 0
    contadorU = 0

    for i in range(0, tamaño):
        nombre = archivo.at[i, "nombre"]

        if nombre[0] == "A":
            contadorA = contadorA + 1
        
        elif nombre[0] == "E":
            contadorE = contadorE + 1

        elif nombre[0] == "I":
            contadorI = contadorI + 1

        elif nombre[0] == "O":
            contadorO = contadorO + 1
        
        elif nombre[0] == "U":
            contadorU = contadorU + 1
        
    print("Hay ", contadorA, "empezados en A,", contadorE, "empezados en E,", contadorI, "empezados en I,", contadorO, "empezados en O,", contadorU, "empezados en U")

def nombreMasUsado():
    masUsado = ""
    cantidad = 0

    for i in range(0, tamaño):
        nombre = archivo.at[i, "nombre"]
        cant = archivo.at[i, "cantidad"]

        if cant > cantidad:
            masUsado = nombre
            cantidad = cant
    
    print("El nombre mas usado es ", masUsado, " con ", cantidad, " de usos")

def nombreMasUsado2014():
    masUsado = ""
    cantidad = 0

    for i in range(0, tamaño):
        nombre = archivo.at[i, "nombre"]
        cant = archivo.at[i, "cantidad"]
        anio = archivo.at[i, "anio"]

        if cant > cantidad and anio == 2014:
            masUsado = nombre
            cantidad = cant
    
    print("En 2014, el nombre mas usado es ", masUsado, " con ", cantidad, " de usos")

def nombreMenosUsado2010():
    menosUsado = ""
    cantidad = 10000

    for i in range(0, tamaño):
        nombre = archivo.at[i, "nombre"]
        cant = archivo.at[i, "cantidad"]
        anio = archivo.at[i, "anio"]

        if cant < cantidad and anio == 2010:
            menosUsado = nombre
            cantidad = cant
    
    print("En 2010, el nombre menos usado es ", menosUsado, " con ", cantidad, " de usos")

def nombreConAMenosUsado2011():
    menosUsado = ""
    cantidad = 10000

    for i in range(0, tamaño):
        nombre = archivo.at[i, "nombre"]
        cant = archivo.at[i, "cantidad"]
        anio = archivo.at[i, "anio"]

        if cant < cantidad and anio == 2011 and nombre[0] == "A":
            menosUsado = nombre
            cantidad = cant
    
    print("En 2011, el nombre menos usado empezado en A es ", menosUsado, " con ", cantidad, " de usos")

def promedioNombresPorAño():
    cantidadTotal2010 = 0
    cantidadNombres2010 = 0
    cantidadTotal2011 = 0
    cantidadNombres2011 = 0
    cantidadTotal2012 = 0
    cantidadNombres2012 = 0
    cantidadTotal2013 = 0
    cantidadNombres2013 = 0
    cantidadTotal2014 = 0
    cantidadNombres2014 = 0

    for i in range(0, tamaño):
        cant = archivo.at[i, "cantidad"]
        anio = archivo.at[i, "anio"]

        if anio == 2010:
            cantidadTotal2010 = cantidadTotal2010 + cant
            cantidadNombres2010 = cantidadNombres2010 + 1
        elif anio == 2011:
            cantidadTotal2011 = cantidadTotal2011 + cant
            cantidadNombres2011 = cantidadNombres2011 + 1
        elif anio == 2012:
            cantidadTotal2012 = cantidadTotal2012 + cant
            cantidadNombres2012 = cantidadNombres2012 + 1
        elif anio == 2013:
            cantidadTotal2013 = cantidadTotal2013 + cant
            cantidadNombres2013 = cantidadNombres2013 + 1
        elif anio == 2014:
            cantidadTotal2014 = cantidadTotal2014 + cant
            cantidadNombres2014 = cantidadNombres2014 + 1

    print("El promedio de nombres utilizados en 2010 es de ", cantidadTotal2010 / cantidadNombres2010, ", en 2011 es de ", cantidadTotal2011 / cantidadNombres2011, ", en 2012 es de ", cantidadTotal2012 / cantidadNombres2012, ", en 2013 es de ", cantidadTotal2013 / cantidadNombres2013, ", en 2014 es de ", cantidadTotal2014 / cantidadNombres2014)

def promedioNombres():
    cantidadTotal = 0
    cantidadNombres = 0

    for i in range(0, tamaño):
        cant = archivo.at[i, "cantidad"]
        
        cantidadTotal = cantidadTotal + cant
        cantidadNombres = cantidadNombres + 1

    print("El promedio de nombres utilizados es de ", cantidadTotal / cantidadNombres)

def sumaNombres():
    cantidadTotal = 0

    for i in range(0, tamaño):
        cant = archivo.at[i, "cantidad"]
        
        cantidadTotal = cantidadTotal + cant

    print("La suma total de nombres utilizados es de ", cantidadTotal)

print("1- ¿Cuántos nombres fueron usados en el año 2012?")
#nombres2012()

print("2- ¿Cuántos nombres fueron usados en el año 2012 y empiezan con la letra S?")
#nombresConS2012()

print("3- ¿Cuántos nombres empiezan con una letra vocal?")
#nombresConVocal()

print("4- ¿Cuántos nombres empiezan con una letra vocal y fueron utilizados desde el año 2010 hasta el año 2013?")
#nombresVocal20102013()

print("5- ¿Cuál es el nombre que fué más utilizado?")
#nombreMasUsado()

print("6- ¿Cuál es el nombre que fué más utilizado en el año 2014?")
#nombreMasUsado2014()

print("7- ¿Cuál fué el nombre menos utilizado en año 2010?")
#nombreMenosUsado2010()

print("8- ¿Cuál fué el nombre menos utilizado en el año 2011 y empieza con la letra A?")
#nombreConAMenosUsado2011()

print("9- ¿Cuál es la cantidad promedio de nombres utilizados?")
#promedioNombres()

print("10- ¿Cuál es la cantidad promedio de nombres utilizados por año?")
#promedioNombresPorAño()

print("11- ¿Cuál es la suma total de los nombres utilizados?")
#sumaNombres()

print("12- ¿Cuántos nombres empiezan con cada una de las vocales?")
#nombresConCadaVocal()

x_values = archivo['nombre'].value_counts().index.tolist()
y_values = archivo['cantidad'].index.tolist()
y_values.reverse()
x_values.reverse()

# gráfico 1
plt1.bar(x_values,y_values)
plt1.title('Cantidad nombres')
plt1.xlabel('nombre')
plt1.ylabel('Cantidad')
plt1.xticks(rotation= 90)
plt1.show()
```
