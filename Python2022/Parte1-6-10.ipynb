```python
import pandas as pd
import matplotlib.pyplot as plt1

archivo = pd.read_csv("actividades-culturales-2021.csv", sep=";")

def presencialVirtual():
    presencial = 0
    virtual = 0

    tamaño = archivo["Modalidad"].size

    for i in range(0, tamaño):

        modalidad = archivo.at[i, 'Modalidad']

        if modalidad == 'Virtual':
            virtual = virtual + 1
        else:
            presencial = presencial + 1

    print("Modalidad presencial: ", presencial, ", y Modalidad virtual: ", virtual)

def lluviaSuspende():
    suspende = 0
    noSuspende = 0

    tamaño = archivo["se_suspende_por_lluvia"].size

    for i in range(0, tamaño):

        modalidad = archivo.at[i, 'se_suspende_por_lluvia']

        if modalidad == 'No':
            noSuspende = noSuspende + 1
        else:
            suspende = suspende + 1

    print("Si llueve, se supenden: ", suspende, " actividades, y no se suspenden: ", noSuspende)

def cine():
    contador = 0
    tamaño = archivo["descripcion_actividad"].size

    for i in range(0, tamaño):
        descripcion = archivo.at[i, 'descripcion_actividad']
        
        if pd.isnull(descripcion):
            descripcion = " "
        
        if "cine" in descripcion:
            contador = contador + 1


    print("Hay ", contador, " actividades relacionadas con el cine")

def actividadesPorBarrio():
    barrios = {}
    tamaño = archivo["barrio"].size

    for i in range(0, tamaño):
        barrio = archivo.at[i, "barrio"]
        array = barrios.keys()
        encontrado = False

        for i in array:
            if i == barrio:
                value = barrios.get(barrio)
                barrios[barrio] = value + 1
                encontrado = True
        
        if encontrado == False:
            barrios.update({barrio: 1})
    
    print(barrios)

def musicaUrquiza():
    contador = 0
    tamaño = archivo["barrio"].size

    for i in range(0, tamaño):
        barrio = archivo.at[i, "barrio"]

        if barrio == "VILLA URQUIZA":
            descripcion = archivo.at[i, 'descripcion_actividad']

            if pd.isnull(descripcion):
                descripcion = " "
        
            if "musica" in descripcion:
                contador = contador + 1
    
    print("Hay ", contador, " actividades relacionadas a la musica en Villa Urquiza")

print("1- ¿Cuántas actividades se realizarán de forma virtual o de forma presencial?")
presencialVirtual()

print("2- ¿Cuántas actividades se supenden por lluvia?")
lluviaSuspende()

print("3- ¿Cuáles son los nombres de las actividades que están relacionadas con el cine?")
cine()

print("4- ¿Cuántas actividades se realizarán por barrio?")
actividadesPorBarrio()

print("5- ¿Cuántas actividades relacionadas con la música sucederán en el barrio de Villa Urquiza?")
musicaUrquiza()

print("El resto de consignas utilizan fechas por lo que no son necesarias")


x_values = archivo['barrio'].value_counts().index.tolist()
y_values = archivo['barrio'].value_counts().tolist()
y_values.reverse()
x_values.reverse()

# gráfico 1
plt1.bar(x_values,y_values)
plt1.title('Cantidad actividades por barrio')
plt1.xlabel('Barrios')
plt1.ylabel('Actividades')
plt1.xticks(rotation= 90)
plt1.show()
```
